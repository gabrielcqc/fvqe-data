# Plot

```python
plt.rcParams.update({
    "text.usetex": True,
    "font.family": "sans-serif",
    "font.sans-serif": "Helvetica",
})

import matplotlib as mpl
import matplotlib.pyplot as plt

mpl.rcParams["font.family"] = "serif"
mpl.rcParams["font.serif"] = "STIX"
mpl.rcParams["mathtext.fontset"] = "stix"
```

```python
def get_subplot(ax1, data, title, l, b, h, w):
    ns = sorted(data[1]['total_min_samples'].keys())
    
    linewidth = 4
    markersize = 12
    
    marks = {1: '-^', 2: ':^', 3: '-v', 4: ':v'}
    marks = {1: '-o', 2: ':o', 3: '-*', 4: ':*'}
    
    colors = {'total_min_samples': '#0571b0', 'total_min_samples_95': '#04617b', 'total_min_samples_9': '#83e4ff'}
    colors = {'total_min_samples': '#8762B3', 'total_min_samples_95': '#C9ACE8', 'total_min_samples_9': '#9ED0E2'}
    # for i in marks:        
    #     ax1.plot([], [], marks[i], c='black', label=f"H.P. {i}", linewidth=linewidth, markersize=markersize)

    # ax1.plot([], [], '-', c=colors['total_min_samples'], label="Ground State", linewidth=linewidth, markersize=markersize)
    # ax1.plot([], [], '-', c=colors['total_min_samples_95'], label=r"App. Rat.$\geq 95$", linewidth=linewidth, 
    #          markersize=markersize)
    # ax1.plot([], [], '-', c=colors['total_min_samples_9'], label=r"App. Rat.$\geq 9$", linewidth=linewidth, 
    #          markersize=markersize)

    # ax1.plot([], [], '-', c='#d7191c', label="4", linewidth=linewidth, markersize=markersize)
    
    for ap in ['total_min_samples', 'total_min_samples_9', 'total_min_samples_95']:
        for i in marks:
            if i in [1, 2]:
                ax1.plot(ns, [np.mean(data[i][ap][n]) for n in ns], marks[i], c=colors[ap], linewidth=linewidth, markersize=markersize, markerfacecolor='none', mew=2.5)
            else:
                ax1.plot(ns, [np.mean(data[i][ap][n]) for n in ns], marks[i], c=colors[ap], linewidth=linewidth, markersize=markersize)
    
    markersize = 15
    ax2 = fig.add_axes([l, b, w, h])
    for i in marks:
        if i in [1, 2]:
            ax2.plot(ns, [len(data[i]['total_min_samples'][n]) for n in ns], marks[i], c=colors['total_min_samples'], linewidth=linewidth-1, markerfacecolor='none', markersize=markersize, mew=2.5)
        else:
            ax2.plot(ns, [len(data[i]['total_min_samples'][n]) for n in ns], marks[i], c=colors['total_min_samples'], linewidth=linewidth-1, markersize=markersize)
    # if title == r'MAXCUT $|$ F-VQE + IQP':
    #     ax1.legend(loc='lower right', fontsize=20)

    ax1.set_yscale('log')
#     ax1.set_title(title, fontsize=20)
#     ax1.set_xlabel(r'Number of qubits', fontsize=20)
#     ax1.set_ylabel(r'Number of samples', fontsize=20)
    
    #ax2.set_ylim(46, 66)
    
    ax2.set_xlabel(r'qubits', fontsize=25)
    ax2.set_title(r'solved problems', fontsize=25)
    ax2.yaxis.set_major_locator(MaxNLocator(integer=True))
    ax2.set_yticks([48, 53, 59, 64])
    ax2.set_ylim(47, 65)
    ax2.set_xticks(ns)
    ax1.set_xticks(ns)
    
    ax2.tick_params(labelsize=23)
    ax1.tick_params(labelsize=25)
    return ax1, ax2
```

```python
fig, axs = plt.subplots(2, 2, figsize=(22,14), dpi=400, sharey=True, sharex=True)

# fig.suptitle(r'Number of samples to obtain a cumulative probability larger than 95%', fontsize=20) # or plt.suptitle('Main title')
l, b, h, w = .074, .795, .16, .185
get_subplot(axs[0,0], data_maxcut_fvqe, r'MAXCUT $|$ F-VQE + IQP', l, b, h, w)
l, b, h, w = .55, .796, .16, .185
get_subplot(axs[0,1], data_maxcut_cl, r'MAXCUT $|$ F-VQE + Classical', l, b, h, w)
l, b, h, w = .074, .329, .16, .185
get_subplot(axs[1,0], data_atsp_fvqe, r'ATSP $|$ F-VQE + IQP', l, b, h, w)
l, b, h, w = .55, .329, .16, .185
get_subplot(axs[1,1], data_atsp_cl, r'ATSP $|$ F-VQE + Classical', l, b, h, w)

linewidth = 4
markersize = 12

marks = {1: '-^', 2: ':^', 3: '-v', 4: ':v'}
marks = {1: '-o', 2: ':o', 3: '-*', 4: ':*'}

colors = {'total_min_samples': '#0571b0', 'total_min_samples_95': '#04617b', 'total_min_samples_9': '#83e4ff'}
colors = {'total_min_samples': '#8762B3', 'total_min_samples_95': '#C9ACE8', 'total_min_samples_9': '#9ED0E2'}


axs[0,0].plot([], [], '-', c=colors['total_min_samples'], label="Ground State", linewidth=linewidth, markersize=markersize)
axs[0,0].plot([], [], '-', c=colors['total_min_samples_95'], label=r"App. Rat.$\geq 95$", linewidth=linewidth, 
            markersize=markersize)
axs[0,0].plot([], [], '-', c=colors['total_min_samples_9'], label=r"App. Rat.$\geq 9$", linewidth=linewidth, 
            markersize=markersize)
axs[0,0].legend(loc='lower right', fontsize=20)

for i in marks:        
    axs[1,0].plot([], [], marks[i], c='black', label=f"Hyperparameters {i}", linewidth=linewidth, markersize=markersize)
axs[1,0].legend(loc='lower right', fontsize=20)

axs[0,0].set_title('MAXCUT + F-VQE', y=1.0, pad=-29, x=0.7, fontsize=28)
axs[0,1].set_title('MAXCUT + F-VQE (Class.)', y=1.0, pad=-29, x=0.68, fontsize=28)
axs[1,0].set_title('ATSP + F-VQE', y=1.0, pad=-29, x=0.7, fontsize=28)
axs[1,1].set_title('ATSP + F-VQE (Class.)', y=1.0, pad=-29, x=0.68, fontsize=28)

axs[1,1].set_xlabel(r'qubits', fontsize=30)
axs[1,0].set_xlabel(r'qubits', fontsize=30)

axs[1,0].set_ylabel(r'samples for success', fontsize=30)
axs[0,0].set_ylabel(r'samples for success', fontsize=30)
fig.tight_layout()
# plt.savefig('hyperparams.pdf', bbox_inches='tight', format='pdf')
# plt.savefig('hyperparams.svg', bbox_inches='tight', format='svg')
plt.show()
# plt.close()
```
